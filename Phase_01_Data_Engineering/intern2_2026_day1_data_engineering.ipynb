# **Mini Project**

```python
import pandas as pd

print("-" * 60)
print(f"{'STUDENT DATA EXPLORER':^60}")
print(f"{'Mini Project Report':^60}")
print("-" * 60)

# =================== 1. PROJECT TITLE DONE ABOVE ===================


# =================== 2. DATASET OVERVIEW ===================
print("\n" + "=" * 50)
print("        DATASET OVERVIEW")
print("=" * 50)

print(f"Total number of students   : {len(df)}")
print(f"Total number of columns    : {len(df.columns)}")
print(f"Number of unique departments: {df['department'].nunique()}")


# =================== 3. DEPARTMENT WISE ANALYSIS ===================
print("\n" + "=" * 50)
print("   DEPARTMENT WISE ANALYSIS")
print("=" * 50)

dept_counts = df["department"].value_counts()
for department, count in dept_counts.items():
    print(f"{department:<20} : {count} students")


# =================== 4. SCORE ANALYSIS ===================
print("\n" + "=" * 50)
print("         SCORE ANALYSIS")
print("=" * 50)

print(f"Highest math score              : {df['math_score'].max():>3}")
print(f"Lowest math score               : {df['math_score'].min():>3}")
print(f"Average math score              : {df['math_score'].mean():5.2f}")
print(f"Average programming score       : {df['programming_score'].mean():5.2f}")
print(f"Average attendance percentage   : {df['attendance_percentage'].mean():5.2f} %")


# =================== 5. GENDER WISE ANALYSIS ===================
print("\n" + "=" * 50)
print("   GENDER WISE ANALYSIS (Average Total Score)")
print("=" * 50)

# Create total_score = math + programming
df["total_score"] = df["math_score"] + df["programming_score"]
gender_avg = df.groupby("gender")["total_score"].mean()

for gender, avg_score in gender_avg.items():
    print(f"{gender:<8} : {avg_score:6.2f}")


# =================== 6. TOP PERFORMERS ===================
print("\n" + "=" * 50)
print("             TOP 5 PERFORMERS")
print("=" * 50)

top_5 = df.sort_values("total_score", ascending=False).head(5)[["name", "department", "total_score"]]
print(f"{'Name':<20} {'Department':<15} {'Total Score'}")
print("-" * 50)
for _, row in top_5.iterrows():
    print(f"{row['name']:<20} {row['department']:<15} {row['total_score']}")


# =================== 7. ATTENDANCE ANALYSIS ===================
print("\n" + "=" * 50)
print("         ATTENDANCE ANALYSIS (Below 75%)")
print("=" * 50)

low_att = df[df["attendance_percentage"] < 75]
if low_att.empty:
    print("All students have attendance above or equal to 75%.")
else:
    print(f"{'Name':<20} {'Department':<15} {'Attendance %'}")
    print("-" * 50)
    for _, row in low_att.iterrows():
        print(f"{row['name']:<20} {row['department']:<15} {row['attendance_percentage']:>10}%")
```
